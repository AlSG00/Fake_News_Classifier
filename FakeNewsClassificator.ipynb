```python
# Connecting libs
import sklearn
import pandas as pd
import numpy as np
```

```python
# Loading and researching dataset
df_news = pd.read_csv("./fake_news.csv")
print(df_news.info())
print(df_news.head(5))
print(df_news.tail(5))
```

```python
# Splitting dataset into train, validate and test one
data_train = df_news[:3000]
data_val = df_news[1000:2000]
data_test = df_news[3000:6000]

data_train.head(10)
```

```python
# Vectorizing our dataset
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
data_train = tfidf_vectorizer.fit_transform(data_train['text'])
data_val = tfidf_vectorizer.fit_transform(data_val['text']) # Возможно придётся убрать
data_test = tfidf_vectorizer.fit_transform(data_test['text'])

```

```python
# Splitting vectorized datasets into features and labels (train and test ones)
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data_train, data_test, train_size=0.3)
```

```python
# Creating and fitting model with passive aggressive classifier
from sklearn.linear_model import PassiveAggressiveClassifier

ml_model = PassiveAggressiveClassifier(C=0.1)
ml_model.fit(x_train, y_train)
```
